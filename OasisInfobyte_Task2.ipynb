```python
# renaming columns
df.columns=['Region', 'Date', 'Frequency', 'Est Unemployment Rate', 'Est Employed', 'Est Labour Participation Rate', 'Area']
df
```

```python
df['Region'].unique()
```

```python
# to see what kinds of information we can extract from given data
print("The region most data of unemployment rate is available for:-")
print(df['Region'].value_counts().idxmax())
```

####Visualization

```python
region_unemployment_sum = df.groupby('Region')['Est Unemployment Rate'].sum()
plt.figure(figsize=(8,10))
plt.pie(region_unemployment_sum, labels=region_unemployment_sum.index, autopct='%1.1f%%', startangle=140)
plt.title("Unemployment Rate based on Region")
plt.show()
```

```python
plt.figure(figsize=(8,10))
area=df['Area'].value_counts()
plt.pie(area, labels=area.index, autopct='%1.1f%%', startangle=140)
plt.title("Unemployment Rate based on Area")
plt.show()
```

```python
# as unemployment rate based on monthly basis can be an important information adding a column of month

import datetime as dt
import calendar

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['month'] =  df['Date'].dt.month

df['Month'] =  df['month'].apply(lambda x: calendar.month_abbr[x])
```

```python
df
```

```python
# as they both don't serve any unique information
df=df.drop(['Frequency', 'month'], axis=1)
df
```

```python
print("The month most data of unemployment rate is available for:-")
print(df['Month'].value_counts().idxmax())
```

```python
month_unemployment_sum = df.groupby('Month')['Est Unemployment Rate'].sum()
plt.figure(figsize=(8,10))
plt.pie(month_unemployment_sum, labels=month_unemployment_sum.index, autopct='%1.1f%%', startangle=140)
plt.title("Unemployment Rate based on Month")
plt.show()
```

```python
plt.figure(figsize=(10,12))
plt.title('Unemployment Rate In India State Wise')
sns.histplot(x='Est Unemployment Rate', hue="Region", data=df)
plt.show()
```

####Encoding and Standardizing Data so it can be used for Training Purpose

```python
labelEncode = LabelEncoder()
df['Area'] = labelEncode.fit_transform(df['Area'])

df
```

```python
columns_to_standardize = ['Est Unemployment Rate', 'Est Employed', 'Est Labour Participation Rate']

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit and transform the selected columns
df[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])
```
